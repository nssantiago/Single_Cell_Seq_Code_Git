```python
def halvingHalfs(sampleName, sampleFilePath):
    filePaths = search_files(sampleFilePath, sampleName)
    
    for filePath in filePaths:
        if 'Half1' in filePath:
            half1 = sc.read(filePath)
            
            half1_A, half1_B, seed = halvingData(half1, None)
            
            # Re-Clustering Half1_A
            # PCA
            sc.tl.pca(half1_A, svd_solver='arpack')

            # Neighbors and UMAP Computations
            sc.pp.neighbors(half1_A, n_neighbors=50, n_pcs =50, use_rep = "Scanorama")
            sc.tl.umap(half1_A) 

            # Re-calculating cluster numbers 
            sc.tl.leiden(half1_A, resolution = 0.3, key_added = "leiden_0.3")

            # Calculates nuclei and gene information specific to the sample object 
            sc.pp.calculate_qc_metrics(half1_A, percent_top=None, log1p=False, inplace=True)
            
            # Re-Clustering Half1_B
            # PCA
            sc.tl.pca(half1_B, svd_solver='arpack')

            # Neighbors and UMAP Computations
            sc.pp.neighbors(half1_B, n_neighbors=50, n_pcs =50, use_rep = "Scanorama")
            sc.tl.umap(half1_B) 

            # Re-calculating cluster numbers 
            sc.tl.leiden(half1_B, resolution = 0.3, key_added = "leiden_0.3")

            # Calculates nuclei and gene information specific to the sample object 
            sc.pp.calculate_qc_metrics(half1_B, percent_top=None, log1p=False, inplace=True)
            
            # Saving Half A and B
            half1_A.write(f'F:/SampleData/SampleQuaters/{sampleName}_Half1_A_Seed{seed}.h5ad')
            half1_B.write(f'F:/SampleData/SampleQuaters/{sampleName}_Half1_B_Seed{seed}.h5ad')
        
        else:
            half2 = sc.read(filePath)
            
            half2_A, half2_B, seed = halvingData(half2, None)
            
            # Re-Clustering Half1_A
            # PCA
            sc.tl.pca(half2_A, svd_solver='arpack')

            # Neighbors and UMAP Computations
            sc.pp.neighbors(half2_A, n_neighbors=50, n_pcs =50, use_rep = "Scanorama")
            sc.tl.umap(half2_A) 

            # Re-calculating cluster numbers 
            sc.tl.leiden(half2_A, resolution = 0.3, key_added = "leiden_0.3")

            # Calculates nuclei and gene information specific to the sample object 
            sc.pp.calculate_qc_metrics(half2_A, percent_top=None, log1p=False, inplace=True)
            
            # Re-Clustering Half1_B
            # PCA
            sc.tl.pca(half2_B, svd_solver='arpack')

            # Neighbors and UMAP Computations
            sc.pp.neighbors(half2_B, n_neighbors=50, n_pcs =50, use_rep = "Scanorama")
            sc.tl.umap(half2_B) 

            # Re-calculating cluster numbers 
            sc.tl.leiden(half2_B, resolution = 0.3, key_added = "leiden_0.3")

            # Calculates nuclei and gene information specific to the sample object 
            sc.pp.calculate_qc_metrics(half2_B, percent_top=None, log1p=False, inplace=True)
            
            half2_A.write(f'F:/SampleData/SampleQuaters/{sampleName}_Half2_A_Seed{seed}.h5ad')
            half2_B.write(f'F:/SampleData/SampleQuaters/{sampleName}_Half2_B_Seed{seed}.h5ad')
```

# Plotting Functions Figure 7

```python
def figure7(sampleObj, sampleObj_half, sampleObj_quarter, sampleObj_eith, sampleInfo, savePath):
    # Create a figure with subplots
    fig = plt.figure(figsize=(20,30), dpi=200)

    row = 6 
    col = 8

    ax1 = plt.subplot2grid((row, col), (0, 0), colspan = 2) # Clustered UMAP Full Sample
    ax2 = plt.subplot2grid((row, col), (0, 2), colspan = 2) # Clustered UMAP 1/2 Sample
    ax3 = plt.subplot2grid((row, col), (0, 4), colspan = 2) # Clustered UMAP 1/4 Sample
    ax4 = plt.subplot2grid((row, col), (0, 6), colspan = 2) # Clustered UMAP 1/8 Sample
    
    ax5 = plt.subplot2grid((row, col), (1, 0), colspan = 2) 
    ax6 = plt.subplot2grid((row, col), (1, 2), colspan = 2)              
    ax7 = plt.subplot2grid((row, col), (1, 4), colspan = 2)          
    ax8 = plt.subplot2grid((row, col), (1, 6), colspan = 2) 
    
    ax9 = plt.subplot2grid((row, col), (2, 0), colspan = 2) 
    ax10 = plt.subplot2grid((row, col), (2, 2), colspan = 2)                       
    ax11 = plt.subplot2grid((row, col), (2, 4), colspan = 2)                 
    ax12 = plt.subplot2grid((row, col), (2, 6), colspan = 2) 

    ax13 = plt.subplot2grid((row, col), (3, 0), colspan = 2) 
    ax14 = plt.subplot2grid((row, col), (3, 2), colspan = 2)                            
    ax15 = plt.subplot2grid((row, col), (3, 4), colspan = 2)                         
    ax16 = plt.subplot2grid((row, col), (3, 6), colspan = 2) 

    ax17 = plt.subplot2grid((row, col), (4, 0), colspan = 2) 
    ax18 = plt.subplot2grid((row, col), (4, 2), colspan = 2)                            
    ax19 = plt.subplot2grid((row, col), (4, 4), colspan = 2)                           
    ax20 = plt.subplot2grid((row, col), (4, 6), colspan = 2) 
                        
    ax21 = plt.subplot2grid((row, col), (5, 0), colspan = 2) 
    ax22 = plt.subplot2grid((row, col), (5, 2), colspan = 2)                         
    ax23 = plt.subplot2grid((row, col), (5, 4), colspan = 2)                          
    ax24 = plt.subplot2grid((row, col), (5, 6), colspan = 2) 
    
    sampleName = sampleInfo[0]
    half = sampleInfo[1]
    quarter = sampleInfo[2]
    eith = sampleInfo[3]
    
    sampleObjLst = [sampleObj, sampleObj_half, sampleObj_quarter, sampleObj_eith]
    geneLst = ['Gad2', 'Pvalb', 'Slc17a7', 'Mbp', 'Gfap']
    axFullSample = [ax5, ax9, ax13, ax17, ax21]
    axHalfSample = [ax6, ax10, ax14, ax18, ax22]
    axFourthSample = [ax7, ax11, ax15, ax19, ax23] 
    axEithSample = [ax8, ax12, ax16, ax20, ax24] 
    axLsts = [axFullSample, axHalfSample, axFourthSample, axEithSample]           
    
    fullSampleNuc = sampleObj.n_obs    
    sc.pl.umap(sampleObj, color='leiden_0.3', legend_loc = 'on data', ax=ax1, show=False)
    ax1.set_title(f'{sampleName} Full Sample ({fullSampleNuc} Nuclei)')

    halfSampleNuc = sampleObj_half.n_obs
    sc.pl.umap(sampleObj_half, color='leiden_0.3', legend_loc = 'on data', ax=ax2, show=False)
    ax2.set_title(f'{sampleName} {half} ({halfSampleNuc} Nuclei)') 

    fourthSampleNuc = sampleObj_quarter.n_obs
    sc.pl.umap(sampleObj_quarter, color='leiden_0.3', legend_loc = 'on data', ax=ax3, show=False)
    ax3.set_title(f'{sampleName} {half} Quarter {quarter} ({fourthSampleNuc} Nuclei)') 

    eithSampleNuc = sampleObj_eith.n_obs       
    sc.pl.umap(sampleObj_eith, color='leiden_0.3', legend_loc = 'on data', ax=ax4, show=False)
    ax4.set_title(f'{sampleName} {half} Quarter {quarter} Eith {eith} ({eithSampleNuc} Nuclei)') 
                           
    for obj, axLst in zip(sampleObjLst, axLsts):
        for gene, axObj in zip(geneLst, axLst):
            sc.pl.umap(obj, color=gene, ax=axObj, show=False)
            axObj.set_title(gene)

    # Adjust the spacing between subplots
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    
    # Saving the figure onto the computer 
    plt.savefig(os.path.join(savePath, f'{sampleName}_{half}_{quarter}_{eith}.jpg'), bbox_inches = 'tight')
```
